```
#Let's import pandas
import pandas as pd
```

```
#Let's grab the dataset
url = 'https://raw.githubusercontent.com/deejacruz/farmers-markets/master/FarmersMarketLatLon.csv'

#Let's set up the dataframe
farmer_df = pd.read_csv(url)
```

```
#Let's check out what types of data we have in each column
farmer_df.dtypes
```

```
#Let's check out the first 10 rows of data
farmer_df.head(10)
```

```
#Let's find out how many rows and columns we have
farmer_df.shape
```

```
#Let's make a scatter plot
farmer_df.plot(kind = "scatter", x = "x", y = "y")
```

```
#Let's import pyplot
import matplotlib.pyplot as plt
```

```
#Let's make a pretty scatter plot that people will actually want to look at
farmer_df.plot(kind = "scatter", x ="x", y="y", alpha=0.5, figsize=(18,10), c='zip', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```

```
#Let's work with a subset focused only on Texas
farmer_texas_df = farmer_df[farmer_df['State'] == 'Texas']

#Let's check out the first ten rows
farmer_texas_df.head(10)
```

```
#Let's check out the shape
farmer_texas_df.shape
```

```
#and verift the data types
farmer_texas_df.dtypes
```

```
#Texas scatter plot
farmer_texas_df.plot(kind = "scatter", x ="x", y="y", alpha=1, figsize=(12,9), c='zip', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```

```
#Let's try Florida
farmer_florida_df = farmer_df[farmer_df['State'] == 'Florida']
```

```
#Florida scatter plot
farmer_florida_df.plot(kind = "scatter", x ="x", y="y", alpha=1, figsize=(10,10), c='zip', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```

```
#Let's look at the Southeast region
southeast = ['Alabama', 'Florida', 'Georgia', 'Arkansas', 'Louisiana', 'Mississippi', 'North Carolina', 'South Carolina', 'Tennessee']
```

```
#Southeast dataframe
farmer_southeast0_df = farmer_df[farmer_df.State.isin(southeast)]
```

```
#Need to drop this FMID
zero = [1011689]
```

```
#Final southeast dataframe
farmer_southeast_df = farmer_southeast0_df[~farmer_southeast0_df.FMID.isin(zero)]
```

```
farmer_southeast_df.shape
```

```
#Southeast scatter plot
farmer_southeast_df.plot(kind = "scatter", x ="x", y="y", alpha=0.7, figsize=(10,8), c='zip', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```
