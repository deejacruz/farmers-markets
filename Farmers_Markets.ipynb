```
#Let's use a different version of the dataset so we can look at WIC and SNAP information
url = 'https://raw.githubusercontent.com/deejacruz/farmers-markets/master/FarmersMarketAssist.csv'

#Create the dataframe
farmer_assist_df = pd.read_csv(url)
```

```
#Check out the data
farmer_assist_df.head(10)
```

```
#Columns and rows
farmer_assist_df.shape
```

```
#Data types
farmer_assist_df.dtypes
```

```
#Dummy code the different types of assistance offered
one_hot_df = pd.get_dummies(farmer_assist_df[['WIC', 'WICcash', 'SFMNP', 'SNAP']])
```

```
#Combine one hot encoding with farmer_assist_df
one_hot_farmer_df = farmer_assist_df[['FMID','city','State','zip','x','y','Credit']].join(one_hot_df)
```

```
#Check the data
one_hot_farmer_df.head(5)
```

```
#Compute score to measure total assistance
one_hot_farmer_df['Score'] = one_hot_farmer_df[['WIC_Y','WICcash_Y','SFMNP_Y','SNAP_Y']].sum(axis=1)
```

```
#Check data
one_hot_farmer_df.head(5)
```

```
#Make a bubbly scatter plot
one_hot_farmer_df.plot(kind = "scatter", x ="x", y="y", alpha=0.7, s=one_hot_farmer_df['Score'], figsize=(18,10), c='Score', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```

```
#Trim the outliers
CONUS = ['Alabama','Arizona','Arkansas','California','Colorado','Connecticut','Delaware','Florida','Georgia','Idaho','Illinois','Indiana','Iowa','Kansas','Kentucky','Louisiana','Maine','Maryland','Massachusetts','Michigan','Minnesota','Mississippi','Missouri','Montana','Nebraska','Nevada','New Hampshire','New Jersey','New Mexico','New York','North Carolina','North Dakota','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island','South Carolina','South Dakota','Tennessee','Texas','Utah','Vermont','Virginia','Washington','West Virginia','Wisconsin','Wyoming']
```

```
#CONUS dataframe
one_hot_farmer_conus0_df = one_hot_farmer_df[one_hot_farmer_df.State.isin(CONUS)]
```

```
#Make a bubbly scatter plot using CONUS
one_hot_farmer_conus0_df.plot(kind = "scatter", x ="x", y="y", alpha=0.7, s=one_hot_farmer_conus0_df['Score'], figsize=(18,10), c='Score', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```

```
#Trim another pesky outlier
outlier = [1019334]
```

```
#Final CONUS
one_hot_farmer_conus_df = one_hot_farmer_conus0_df[~one_hot_farmer_conus0_df.FMID.isin(outlier)]
```

```
#Final bubbly CONUS plot showing farmers markets in the contiguous U.S. where a higher score (coloer red) indicates markets that accept all four food assistance programs as forms of payment
one_hot_farmer_conus_df.plot(kind = "scatter", x ="x", y="y", alpha=0.7, s=one_hot_farmer_conus_df['Score']*10, figsize=(18,10), c='Score', cmap = plt.get_cmap('rainbow'), colorbar=True)
plt.legend
```
